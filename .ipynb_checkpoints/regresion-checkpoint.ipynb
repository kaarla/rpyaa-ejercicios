```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from semana2_datos import *

X_1 = np.array([[1,x] for x, y in data_1])
Y_1 = np.array([y for x, y in data_1])

X_2 = np.array([[1,x] for x, y in data_2])
Y_2 = np.array([y for x, y in data_2])

X_3 = np.array([[1,x] for x, y in data_3])
Y_3 = np.array([y for x, y in data_3])

X_4 = np.array([[1,x] for x, y in data_4])
Y_4 = np.array([y for x, y in data_4])

# data_1
x_1 = np.array([x for _, x in X_1])
plt.scatter(x_1, Y_1)
plt.show()

# data_2
x_2 = np.array([x for _, x in X_2])
plt.scatter(x_2, Y_2)
plt.show()

# data_3
x_3 = np.array([x for _, x in X_3])
plt.scatter(x_3, Y_3)

# data_4
x_4 = np.array([x for _, x in X_4])
plt.scatter(x_4, Y_4)
```

```python
# data_1
W_1 = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_1.T,
                                                  X_1)), X_1.T), Y_1)
f1 = lambda X: W_1[1]*X + W_1[0]
print('Los pesos aprendidos para data_1 son\nw0: {}\nw1: {}\n'.format(W_1[0], W_1[1]))

# data_2
W_2 = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_2.T,
                                                  X_2)), X_2.T), Y_2)
f2 = lambda X: W_2[1]*X + W_2[0]
print('Los pesos aprendidos para data_2 son\nw0: {}\nw1: {}\n'.format(W_2[0], W_2[1]))

# data_3
W_3 = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_3.T,
                                                  X_3)), X_3.T), Y_3)
f3 = lambda X: W_3[1]*X + W_3[0]
print('Los pesos aprendidos para data_3 son\nw0: {}\nw1: {}\n'.format(W_3[0], W_3[1]))

# data_4
W_4 = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_4.T,
                                                  X_4)), X_4.T), Y_4)
f4 = lambda X: W_4[1]*X + W_4[0]
print('Los pesos aprendidos para data_4 son\nw0: {}\nw1: {}\n'.format(W_4[0], W_4[1]))
```

```python
# data_1
plt.scatter(x_1, Y_1)
lower = min(x_1)
upper = max(x_1)
sample_x1 = np.linspace(lower, upper, num=len(x_1))
plt.plot(sample_x1, f1(sample_x1), color='r')
plt.show()

# data_2
plt.scatter(x_2, Y_2)
lower = min(x_2)
upper = max(x_2)
sample_x2 = np.linspace(lower, upper, num=len(x_2))
plt.plot(sample_x2, f2(sample_x2), color='g')
plt.show()

# data_3
plt.scatter(x_3, Y_3)
lower = min(x_3)
upper = max(x_3)
sample_x3 = np.linspace(lower, upper, num=len(x_3))
plt.plot(sample_x3, f3(sample_x3), color='r')

# data_3
plt.scatter(x_4,Y_4)
lower = min(x_4)
upper = max(x_4)
sample_x4= np.linspace(lower, upper, num=len(x_4))
plt.plot(sample_x4, f4(sample_x4), color='g')
```

```python
# Polynomial regression
x_1 = np.expand_dims(x_1, axis=1)
poly = PolynomialFeatures(3)
features = poly.fit_transform(x_1)

lin_reg = LinearRegression()
lin_reg.fit(features, Y_1)

sample_x1 = np.expand_dims(sample_x1, axis=1)
sample_x1_poly = poly.fit_transform(sample_x1)
y1_poly = lin_reg.predict(sample_x1_poly)
plt.scatter(x_1, Y_1)
plt.plot(sample_x1, y1_poly, color='g')
```

```python
# Polynomial regression
x_2 = np.expand_dims(x_2, axis=1)
poly = PolynomialFeatures(3)
features = poly.fit_transform(x_2)

lin_reg = LinearRegression()
lin_reg.fit(features, Y_2)

sample_x2 = np.expand_dims(sample_x2, axis=1)
sample_x2_poly = poly.fit_transform(sample_x2)
y2_poly = lin_reg.predict(sample_x2_poly)
plt.scatter(x_2, Y_2)
plt.plot(sample_x2, y2_poly, color='g')
```

```python
# Polynomial regression
x_3 = np.expand_dims(x_3, axis=1)
poly = PolynomialFeatures(3)
features = poly.fit_transform(x_3)

lin_reg_3 = LinearRegression()
lin_reg_3.fit(features, Y_3)

sample_x3 = np.expand_dims(sample_x3, axis=1)
sample_x3_poly = poly.fit_transform(sample_x3)
y3_poly = lin_reg_3.predict(sample_x3_poly)
plt.scatter(x_3, Y_3)
plt.plot(sample_x3, y3_poly, color='g')

# Polynomial regression
x_4 = np.expand_dims(x_4, axis=1)
poly = PolynomialFeatures(3)
features = poly.fit_transform(x_4)

lin_reg_4 = LinearRegression()
lin_reg_4.fit(features, Y_4)

sample_x4 = np.expand_dims(sample_x4, axis=1)
sample_x4_poly = poly.fit_transform(sample_x4)
y4_poly = lin_reg_4.predict(sample_x4_poly)
plt.scatter(x_4, Y_4)
plt.plot(sample_x4, y4_poly, color='y')
```

```python
# Ridge para x_1
ridge = Ridge(alpha=2.5)
ridge.fit(features, Y_1)

y1_ridge = ridge.predict(sample_x1_poly)
plt.scatter(x_1, Y_1)
plt.plot(sample_x1, y1_ridge, color='r')
```

```python
# Ridge para x_2
ridge = Ridge(alpha=2.5)
ridge.fit(features, Y_2)

y2_ridge = ridge.predict(sample_x2_poly)
plt.scatter(x_2, Y_2)
plt.plot(sample_x2, y2_ridge, color='r')
```

```python
# Ridge para x_3 y x_4
ridge = Ridge(alpha=2.5)
ridge.fit(features, Y_3)

y3_ridge = ridge.predict(sample_x3_poly)
plt.scatter(x_3, Y_3)
plt.plot(sample_x3, y3_ridge, color='r')

ridge = Ridge(alpha=2.5)
ridge.fit(features, Y_4)

y4_ridge = ridge.predict(sample_x4_poly)
plt.scatter(x_4, Y_4)
plt.plot(sample_x4, y4_ridge, color='y')
```

```python
# Lasso para x_1
from sklearn import linear_model
lasso = linear_model.Lasso(alpha=2.5)
lasso.fit(features, Y_1)

y1_lasso = lasso.predict(sample_x1_poly)
plt.scatter(x_1, Y_1)
plt.plot(sample_x1, y1_lasso, color='r')
```

```python
# Lasso para x_2

lasso = linear_model.Lasso(alpha=2.5)
lasso.fit(features, Y_2)

y2_lasso = lasso.predict(sample_x1_poly)
plt.scatter(x_1, Y_1)
plt.plot(sample_x1, y2_lasso, color='r')
```

```python
# Lasso para x_3 y x_4

lasso = linear_model.Lasso(alpha=2.5)
lasso.fit(features, Y_3)

y3_lasso = lasso.predict(sample_x3_poly)
plt.scatter(x_3, Y_3)
plt.plot(sample_x3, y3_ridge, color='r')

lasso = linear_model.Lasso(alpha=2.5)
lasso.fit(features, Y_4)

y4_lasso = lasso.predict(sample_x4_poly)
plt.scatter(x_4, Y_4)
plt.plot(sample_x4, y4_ridge, color='y')
```
